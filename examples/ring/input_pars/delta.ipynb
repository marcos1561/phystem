```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
def load_data(path):
    p = Path(path)
    deltas = np.load(p/"deltas.npy")
    times = np.load(p/"times.npy")
    return deltas, times
```

```python
path = "data/align_1/delta/data"

deltas, times = load_data(path)
times = times[:-1]

plt.xlabel("t")
plt.ylabel("$\\langle \\Delta \\rangle$")

plt.plot(times, deltas, "-o")
```
