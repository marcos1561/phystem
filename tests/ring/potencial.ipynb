```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def leonard(r, epsilon, sigma):
    y = epsilon/12 * ((sigma/r)**12 - (sigma/r)**6)
    y = epsilon/12 * (12 * (sigma/r)**11 * sigma * (-1) / r**2 - 6 * (sigma/r)**5 * sigma * (-1) / r**2)
    # y[r>(2**(1/6)*sigma)] = 0
    return y

def lines(r, fr, fa, re, ro):
    r_filter = r <= re
    a_filter = np.logical_and((re < r), (r <= ro))

    y = np.zeros(r.size)
    y[r_filter] = fr * (r[r_filter] - re)/(re)
    y[a_filter] = fa * (r[a_filter] - re)/(ro - re)
    
    return y
```

```python
sigma = 1
epsilon = 1

fr = 30
fa = 0.75
re = 1
ro = 1.1

x = np.linspace(0.8, 2, 1000)

plt.plot(x, leonard(x, epsilon, sigma))
plt.plot(x, lines(x, fr, fa, re, ro))
```
