```python
fig, ax = plt.subplots()
ax: Axes

x = [1, 2, 3]
y = [1, 2, 3]
u = np.array([1, 2, 3])
v = np.array([1, 1, 1])

arrows = ax.quiver(x, y, u, v)

arrows.set_UVC(V=-v, U=u)
arrows.set_offsets([
    [x[0], y[0]],
    [x[1], y[1]+0.5],
    [x[2], y[2]],
])

# arrows.set_offsets([])
# arrows.set_UVC(V=[], U=[])
arrows.remove()
# arrows.remove()
# ax.add_artist(arrows)


ax.set_xlim(right=3.5)
plt.show()
```
